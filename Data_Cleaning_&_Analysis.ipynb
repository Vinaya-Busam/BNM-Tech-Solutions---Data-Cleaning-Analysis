## Importing Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
```

## Load & Understand the Dataset

```python
dataset = pd.read_csv("/content/netflix_titles.csv")
```

## Handling Missing Values

#### Cheking missing values

```python
dataset.isnull().sum()
```

#### Filling missing values

```python
dataset['director'].fillna('Unknown', inplace=True)
dataset['cast'].fillna('Unknown', inplace=True)
dataset['country'].fillna('Unknown', inplace=True)

dataset['rating'].fillna(dataset['rating'].mode()[0], inplace=True)
```

```python
dataset.dropna(subset=['date_added', 'duration'], inplace=True)
```

```python
dataset.isnull().sum()
```

```python
dataset.head()
```

```python
print("Shape of dataset after handling missing values:")
dataset.shape
```

# Remove Duplicate Records

```python
duplicates = dataset.duplicated().sum()
print("Duplicated Rows:", duplicates)
```

There are no duplicated records

# Data Cleaning & Preprocessing

#### Convert Date Column

```python
dataset['date_added'] = pd.to_datetime(dataset['date_added'], format='mixed')
```

```python
dataset.head()
```

#### Extract Year Added

```python
dataset['year_added'] = dataset['date_added'].dt.year
```

#### Clean duration column


```python
dataset['duration_num'] = dataset['duration'].str.extract(r'(\d+)')
dataset['duration_num'] = pd.to_numeric(dataset['duration_num'])
```

```python
dataset.head()
```

#### Remove Extra Spaces

```python
dataset['country'] = dataset['country'].str.strip()
dataset['director'] = dataset['director'].str.strip()
```

```python
dataset.head()
```
